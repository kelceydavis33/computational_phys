a) Write a program that reads in the data and makes a graph of sunspots as a function of

```python
def read_sunspots(path):
    """
    Reads in a file containing information about sunspot data and produces a plot of the sunspots over time
    
    Parameters
    ---
    path: string, absolute path to the sunspot file
    
    Returns
    ---
    Produces a matplotlib plot of the sunspots over time. Sunspot count is the number of sunspots observed
    in a given month since 01-01-1749.

    """
    #Define an empty list to store the dates
    dates = []
    #Read in the file as a pandas DataFrame, indicate that the file contains to headers
    df = pd.read_table(path, header = None)
    #Months are indicated in the first column and spot counts are indicated in the second
    months = df[0].values
    spots = np.array(df[1].values)
    #Now I begin constructing a list of astropy Time objects to store the dates that these sunspots were 
    #reported. I know that you probably just want me to plot the spots vs months as they are, but 
    #this would technically not be correct since each month has a different lenght. To account for 
    #month length variation and leap years, the easiest way to produce the correct plot is to use astropy
    #Time objects. 
    #Initialize the first date at January 1st 1749
    string_time = "1749-01-01"
    #Add this to the list of date strings
    dates.append(string_time)
    #Loop through the a range the same length as the number of months minus the initialized month
    for month in range(len(months)-1):
        #Grab the month and the year from the string_time, assume data is always reported on the 1st of the month
        m = int(string_time.split('-')[1])
        yr = int(string_time.split('-')[0])
        #If the month is December, increase the year by one and reset the month to January
        if m == 12:
            yr += 1
            m = 1
            #If the month is of a single digit,add a 0 to the front of it in the string
            if len(str(m)) == 1:
                string_time = f"{yr}-0{m}-01"
                dates.append(string_time)
            #If the month is two digits, do not add an extra zero.
            else:
                string_time = f"{yr}-{m}-01"
                dates.append(string_time)
        #if the month is not december, only increase the month by one.
        else:
            m+=1
            if len(str(m)) == 1:
                string_time = f"{yr}-0{m}-01"
                dates.append(string_time)
            else:
                string_time = f"{yr}-{m}-01"
                dates.append(string_time)
    #Set the figure size to be 20x10
    plt.rcParams['figure.figsize'] = (20, 10)
    #initialize matplotlib Time object plotting support
    time_support()
    #Plot the dates vs the sunspots
    plt.plot(Time(dates, format = 'iso'), spots)
    #Rotate the axis ticks so the dates are easier to read
    plt.xticks(rotation = 45)
    #Add minor ticks to the x axis
    plt.tick_params('x', length=10, width=1, which='minor')
    #Add a title and x and y labels
    plt.title('Sunspots over time')
    plt.xlabel('Date')
    plt.ylabel('Sunspot Count');
```

c) Modify your program further to calculate and plot the running average of the data, de-
fined by

$Y_k = \frac{1}{2r}  \sum_{m=-r}^{r} y_{k+m}  $


where r = 5 in this case (and the yk are the sunspot numbers). Have the program plot

```python
def read_sunspots_running_average_first_1000(path):
    """
    Reads in a file containing information about sunspot data and produces a plot of the sunspots over time
    
    Parameters
    ---
    path: string, absolute path to the sunspot file
    
    Returns
    ---
    Produces a matplotlib plot of the sunspots over time. Sunspot count is the number of sunspots observed
    in a given month since 01-01-1749.
    """
    #Define an empty list to store the dates
    dates = []
    #Read in the file as a pandas DataFrame, indicate that the file contains to headers
    df = pd.read_table(path, header = None)
    #Months are indicated in the first column and spot counts are indicated in the second
    months = df[0].values[0:1000]
    spots = np.array(df[1].values)[0:1000]
    #Initialize the first date at January 1st 1749
    string_time = "1749-01-01"
    #Add this to the list of date strings
    dates.append(string_time)
    #Define an empty list to hold the running averages
    yks = []
    #Define the r variable
    r=5
    #Loop through each position in the sunspots data
    for pos in range(len(spots)):
        #Create an interval for the average of +/- 5 data points around the current position
        data_interval = spots[pos-r:pos+r]
        #compute the yk value and append it to the running list of vlaues for the plot
        yks.append(np.sum(data_interval)*(1/(2*r)))
    #Loop through the a range the same length as the number of months minus the initialized month
    for month in range(len(months)-1):
        #Grab the month and the year from the string_time, assume data is always reported on the 1st of the month
        m = int(string_time.split('-')[1])
        yr = int(string_time.split('-')[0])
        #If the month is December, increase the year by one and reset the month to January
        if m == 12:
            yr += 1
            m = 1
            #If the month is of a single digit,add a 0 to the front of it in the string
            if len(str(m)) == 1:
                string_time = f"{yr}-0{m}-01"
                dates.append(string_time)
            #If the month is two digits, do not add an extra zero.
            else:
                string_time = f"{yr}-{m}-01"
                dates.append(string_time)
         #if the month is not december, only increase the month by one.
        else:
            m+=1
            if len(str(m)) == 1:
                string_time = f"{yr}-0{m}-01"
                dates.append(string_time)
            else:
                string_time = f"{yr}-{m}-01"
                dates.append(string_time)
    #Set the figure size to be 20x10
    plt.rcParams['figure.figsize'] = (20, 10)
    #initialize matplotlib Time object plotting support
    time_support()
    #Plot the dates vs the sunspots
    plt.plot(Time(dates, format = 'iso'), spots, label = 'Raw Data')
    #Plot the running average of the sunspots
    plt.plot(Time(dates, format = 'iso'), yks, label = 'Running Average', c = 'k', lw = 3)
    #Rotate the axis ticks so the dates are easier to read
    plt.xticks(rotation = 45)
    #Add minor ticks to the x axis
    plt.tick_params('x', length=10, width=1, which='minor')
    #Add a title and x and y labels
    plt.title('Sunspots over time, fist 1000 points')
    plt.xlabel('Date')
    plt.ylabel('Sunspot Count')
    #Display the legend
    plt.legend();
```

Problem 3: The Mandelbrot set


The Mandelbrot set, named after its discoverer, the French mathematician Benot Mandelbrot,
is a fractal, an infinitely ramified mathematical object that contains structure within structure
within structure, as deep as we care to look. The definition of the Mandelbrot set is in terms of
complex numbers as follows.


Consider the equation
$z = z^2 + c$


where z is a complex number and c is a complex constant. For any given value of c this equa-
tion turns an input number z into an output number z. The definition of the Mandelbrot set
involves the repeated iteration of this equation: we take an initial starting value of z and feed
it into the equation to get a new value z. Then we take that value and feed it in again to get
another value, and so forth. 

The Mandelbrot set is the set of points in the complex plane that
satisfies the following definition:


For a given complex value of c, start with z = 0 and iterate repeatedly. If the magnitude |z|
of the resulting value is ever greater than 2, then the point in the complex plane at position c
is not in the Mandelbrot set, otherwise it is in the set.


In order to use this definition one would, in principle, have to iterate infinitely many times
to prove that a point is in the Mandelbrot set, since a point is in the set only if the iteration
never passes |z| = 2 ever. In practice, however, one usually just performs some large number
of iterations, say 100, and if |z| hasnt exceeded 2 by that point then we call that good enough.



Write a program to make an image of the Mandelbrot set by performing the iteration for all
values of c = x + iy on an N  N grid spanning the region where 2  x  2 and 2  y 2


Make a density plot in which grid points inside the Mandelbrot set are colored black and
those outside are colored white. The Mandelbrot set has a very distinctive shape that looks
something like a beetle with a long snoutyoull know it when you see it.


Hint: You will probably find it useful to start off with quite a coarse grid, i.e., with a small
value of Nperhaps N = 100so that your program runs quickly while you are testing it.
Once you are sure it is working correctly, increase the value of N to produce a final high-quality
image of the shape of the set.
If you are feeling enthusiastic, here is another variant of the same exercise that can produce
amazing looking pictures. Instead of coloring points just black or white, color points according
to the number of iterations of the equation before |z| becomes greater than 2 (or the maximum
number of iterations if |z| never becomes greater than 2). If you use one of the more colorful
color schemes Python provides for density plots, such as the hot or jet schemes, you can
make some spectacular images this way. Another interesting variant is to color according to the
logarithm of the number of iterations, which helps reveal some of the finer structure outside
the set

```python
#Pixels
N = 1000
#Allowed Iterations
n = 1000
#Define xspan and yspan to cover axes from -2 to 2 and contain N pixels between these values
xspan,yspan == np.linspace(-2,2,N)
#Define an empty matrix to store the data. It should have dimmensions of the number of pixels by the number 
#of pixels
matrix = np.zeros([N,N])
#enumerate the xspan and yspan variables so we can loop through each position in the matrix. xi and yi
#are the indices and x and y are the respective x and y values in the matrix
for xi, x in enumerate(xspan):
    for yi, y in enumerate(yspan):
        #Initialize z to be 0
        z = complex(0,0)
        #Loop from 0 to the maximum allowed iterations
        for i in range(0,n):
            #Define c to be a complex number indicated by the position on the x and y axes
            c = complex(x,y)
            #Increase z by z squared and c
            z = (z**2) + c
            #If at any point the absolute value of z is greater than 2, break the loop and add 1 to the 
            #x and y position in the matrix
            if abs(z)<2:
                continue
            else:
                matrix[xi,yi] +=1
                break
```

```python
#Use imshow to display the matrix. Set the origin to be the lower right corner and define the extenet to
#indicate the range over which x and y were defined
plt.imshow(matrix, origin = 'lower', cmap = 'gray', extent=[-2,2,-2,2])
#Display the colorbar
plt.colorbar()
#Add x and y labels
plt.xlabel('x')
plt.ylabel('y')
#Add a title
plt.title('The Mandelbrot Set');
```

```python
def f(x):
    """
    Takes an input x value and returns a y value for the defined function
    
    Parameters:
    ---
    x: input
    
    Returns
    ---
    y: output of function y = x*(x-1)
    
    """
    #Computes the function value at the given x and returns it
    y = (x*(x-1))
    return y 
```

```python
def derrivative(function, x, delta):
    """
    Takes the derrivative of a given function
    
    Parameters
    ---
    function: a function to be evaluated for the derrivative
    
    x: float, x position to evaluate the derrivative at
    
    delta: float, incrimental area around which the derrivative will be computed
    
    Returns
    ---
    value of the specified derrivative
    """
    return(((f(x+delta) - f(x))/(delta)))
```

```python
deriv = derrivative(f,x = 1, delta = 10**-2)
print(f'The derrivative at x=1 for x(x-1) is {deriv}')
```
