```python
from gaussxw import *
def get_gauss(N, a, b, func, t):
    #GAUSSXW FUNCTION CALL FROM CLASS 9/21
    ###
    # calculate the sample points and weights
    x, w = gaussxw(N)
    # map the sample points and weights to the required domain
    xp = 0.5*(b-a)*x + 0.5*(b+a)
    wp = 0.5*(b-a)*w
    # perform the integral
    s = 0.0
    for k in range(N):
        s += wp[k] * func(xp[k], t)
    ###
    return s
```

## Runge-Kutta method

* calculate slope at t+h/2


```python
def f(x,t):
    return (-1*(x**3) + np.sin(t))
NS = [10, 20, 50, 100]
for N in NS:
    a = 0
    b = 10
    #N = 1000
    h = (b-a)/N
    x = 0

    tpoints = np.arange(a,b, h)
    xpoints = []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f((x + 0.5*k1),(t+0.5*h))
        x+= k2
    plt.plot(tpoints, xpoints)
    plt.title(f'N = {N}')
    plt.figure();
```

# 4th order r-k

```python
def f(x,t):
    return (-1*(x**3) + np.sin(t))
NS = [10, 20, 50, 100]
for N in NS:
    a = 0
    b = 10
    #N = 1000
    h = (b-a)/N
    x = 0

    tpoints = np.arange(a,b, h)
    xpoints = []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f((x + 0.5*k1),(t+0.5*h))
        k3 = h*f((x + 0.5*k2),(t+0.5*h))
        k4 = h*f((x + k3),(t+h))
        x+= k4
    plt.plot(tpoints, xpoints)
    plt.title(f'N = {N}, 4th order approx.')
    plt.figure();
```
